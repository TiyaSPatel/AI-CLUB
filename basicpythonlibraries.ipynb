```python
import pandas as pd
url =  "https://archive.ics.uci.edu/ml/machine-learning-databases/iris/iris.data"
columns =  ["sepal_length", "sepal_width", "petal_length", "petal_width", "species"]
df = pd.read_csv(url, header=None, names=columns)
print(df.describe)
print("\nBasic Data Properties:")
print("Number of Rows and Columns:")
print(df.shape)
```

```python
print("columns:")
print(df.columns)
print("datatype")
print(df.dtypes)
```

```python
print("\nFiltering Data:")
verginica_df = df[df['species'] == 'Iris-verginica']
print("Setosa Species Data:")
print(setosa_df.head())
```

```python
print("\nSorting Data:")
sorted_df = df.sort_values(by='sepal_length', ascending=True)
print("Sorted Data by Sepal Length:")
print(sorted_df.head())
```

```python
missing_data = df.isnull().sum()
print("\nMissing Data:")
print(missing_data)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt


# Use Seaborn to create advanced plots for the Iris dataset
sns.set(style="whitegrid")

# Pair plot for visualizing relationships between features
sns.pairplot(df, hue="species", markers=["o", "s", "D"], height=3)
plt.suptitle("Pair Plot of Iris Dataset")
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.boxplot(x="species", y="petal_width", data=df, palette="pastel")
plt.title("Box Plot of Petal Width by Species")
plt.show()
```
